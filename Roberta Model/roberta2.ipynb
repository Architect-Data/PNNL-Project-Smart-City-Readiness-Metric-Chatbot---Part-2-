```python
# Import necessary libraries
from transformers import pipeline , RobertaTokenizer, RobertaForQuestionAnswering
import pdfplumber
import torch
```

```python
# Define the name of the pre-trained model to be used for question answering.
model_name = "deepset/roberta-base-squad2"
# Load the pre-trained model for question answering and move it to the GPU if available.
model = RobertaForQuestionAnswering.from_pretrained(model_name).cuda()
# Load the tokenizer associated with the pre-trained model and set it to use the GPU.
tokenizer = RobertaTokenizer.from_pretrained(model_name, use_cuda=True)
# Determine whether to use the GPU for computation, if available; otherwise, use the CPU.
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
#Load the questions to the pipeline
qa_pipeline = pipeline("question-answering", model=model, tokenizer=tokenizer, device=device)
```

```python
#for individual pdf's of cities, use this to read the text
def read_pdf_text(file_path):
    with pdfplumber.open(file_path) as pdf:
        pdf_text = ""
        for page in pdf.pages:
            pdf_text += page.extract_text()
    return pdf_text


pdf_file_path = "/content/Paris-Resilience-Strategy-English.pdf"
pdf_text = read_pdf_text(pdf_file_path)

#pdf_file_path1 = "/content/Pune-Resilience-Strategy-English.pdf"
#pdf_text1 = read_pdf_text(pdf_file_path1)

#pdf_file_path2 = "/content/Pune-Resilience-Strategy-English.pdf"
#pdf_text2 = read_pdf_text(pdf_file_path2)
```

```python
#for multiple pdf's of cities, use this to read the text
import os
import pdfplumber

def read_pdf_text(file_path):
    with pdfplumber.open(file_path) as pdf:
        pdf_text = ""
        for page in pdf.pages:
            pdf_text += page.extract_text()
    return pdf_text

# Directory containing PDF files
pdf_directory = "/content/"

# Initialize a variable to store combined text
pdf_text = ""

# Iterate over PDF files in the directory
for pdf_file in os.listdir(pdf_directory):
    if pdf_file.endswith(".pdf"):
        pdf_path = os.path.join(pdf_directory, pdf_file)
        pdf_text += read_pdf_text(pdf_path)

# Save combined text as a single variable
combined_text_file_path = "/content/combined_text.txt"
with open(combined_text_file_path, "w") as text_file:
    text_file.write(pdf_text)
```

```python
#Define the question
question = "What is the major concern of Paris?"
```

```python
#Define the answer alongside confidence
result = qa_pipeline(question=question, context=pdf_text)
answer = result["answer"]
confidence = result["score"]
print(answer)
print(confidence)
```
